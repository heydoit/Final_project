```python
import pandas as pd
from soyspacing.countbase import CountSpace
from soynlp.normalizer import *
import urllib.request
import soyspacing
import re

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/증권데이터/herald_2columns_final_data.csv',lineterminator='\n')
```

```python
# 이메일 제거, 

def text_preprocessor(s):

    pattern = r"[a-zA-Z0-9+-]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+$" #이메일 제거
    s = re.sub(pattern=pattern, repl='', string=s)

    s = s.lower() # 대문자를 소문자로 변환
    
    # [], (), {}, <> 괄호와 괄호 안 문자 제거하기
    pattern = r'\([^)]*\)'  # ()
    s = re.sub(pattern=pattern, repl='', string=s)
    
    pattern = r'\[[^)]*\]'  # []
    s = re.sub(pattern=pattern, repl='', string=s)
    
    pattern = r'\<[^)]*\>'  # <>
    s = re.sub(pattern=pattern, repl='', string=s)
    
    pattern = r'\{[^)]*\}'  # {}
    s = re.sub(pattern=pattern, repl='', string=s)
        
    # 특수문자, 숫자 제거 + 띄어쓰기로 대체 X
    pattern = r'[^a-zA-Z가-힣]'
    s = re.sub(pattern=pattern, repl=' ', string=s)

    # OOO 기자 제거
    s = re.sub(r'... 기자','',s)
    
    # 단위 제거 + 숫자 빠지고 난 뒤에 혼자 남은 애매한 단어들까지 제거
    units = ['mm', 'cm', 'km', 'ml', 'kg', 'g',' 년 ',' 월 ',' 일 ',' 사진 ']
    for unit in units:
        s = s.replace(unit, '')

    return s
```
